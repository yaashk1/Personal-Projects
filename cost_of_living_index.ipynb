# Cost of Living Index

##Load Data

```python
import pandas as pd
df = pd.read_csv('/content/Cost_of_Living_Index_by_Country_2024.csv')
df
```

###Data preparation

```python
y = df['Cost of Living Index']
y
```

```python
X = df.drop('Cost of Living Index', axis = 1)
X = df.drop('Country', axis = 1)
X
```

### Data Splitting
split data 80 percent training, 20 percent testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

#Model building

###Linear Regression

Training the model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

Applying the model to make a prediction

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

**Evaluate model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MRE (TRAIN): ' , lr_train_mse)
print('LR R2 (TRAIN): ' , lr_train_r2)
print('LR MRE (TEST): ' , lr_test_mse)
print('LR R2 (TEST): ' , lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

#Hyperparameter Tuning
Hyperparameter tuning is a crucial step in machine learning where we systematically explore different combinations of model settings (hyperparameters) to find the configuration that yields the best performance

```python
from sklearn.model_selection import GridSearchCV

from sklearn.ensemble import RandomForestRegressor # Import the RandomForestRegressor


param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf = RandomForestRegressor(random_state=100) # Create an instance of the RandomForestRegressor

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

best_rf = RandomForestRegressor(**best_params, random_state=100)
best_rf.fit(X_train, y_train)

y_best_rf_train_pred = best_rf.predict(X_train)
y_best_rf_test_pred = best_rf.predict(X_test)
```
